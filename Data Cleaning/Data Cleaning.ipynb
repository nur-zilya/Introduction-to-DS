```python
df_clean = df.copy()
df_clean = df_clean.drop(columns=['county'])
```

```python
def print_useful_rows_info(df): 

   print('Количество полностью заполненных объектов из всей     выборки: ', len(df.dropna()))

   print('Процент полностью заполненных объектов из всей выборки: ', round(len(df.dropna()) / len(df) * 100, 2))
```

```python
print_useful_rows_info(df_clean)
```

```python
df_clean = df_clean.drop(columns=['VIN', 'condition', 'cylinders', 'size', 'drive', 'paint_color', 'type'])
```

```python
df_clean.manufacturer = df_clean.manufacturer.fillna('other')
```

```python
df_clean.year.isna().sum()
```

```python
df_clean.year
```

```python
median_year = df_clean['year'].median()
median_year
```

```python
df_clean[df_clean['year'].isna()] 
```

```python
df_clean['year'] = df_clean.year.fillna(median_year)
df_clean['year'] 
```

```python
df_clean.fuel = df_clean.fuel.fillna('other')
```

```python
df_clean['model'] = df_clean.groupby('manufacturer')['model'].transform(lambda x: x.fillna(x.mode()[0]))
df_clean.model.isna().sum()
```

```python
df_clean.dtypes
```

```python
df_clean['year'] = df_clean['year'].astype(int)
```

```python
categorical_columns = [
    'manufacturer', 'model', 'fuel', 'title_status',
    'transmission', 'state', 'price_category'
]

for col in categorical_columns:
    df_clean[col] = df_clean[col].astype('category')
```

```python
pd.to_datetime(df_clean.posting_date, utc=True)
```

```python
# Найти пороги для отнесения значения к выбросам
def calculate_outliers(data):
    q25 = data.quantile(0.25)
    q75 = data.quantile(0.75)
    iqr = q75 - q25
    boundaries = (q25 - 1.5 * iqr, q75 + 1.5 * iqr)
    return boundaries
```

```python
boundaries = calculate_outliers(df_clean.year)
```

```python
is_outlier = (df_clean['year'] < boundaries[0]) | (df_clean['year'] > boundaries[1])
```

```python
print(f"Количество выбросов {is_outlier.sum()  }")
print(f"Доля выбросов {is_outlier.sum() / len(df_clean)}")
```
